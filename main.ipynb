```python
import re

def extract_filters(query):
    # Extract city (words like Pune, Mumbai, etc.)
    city_match = re.search(r'in\s+([a-zA-Z\s]+)', query)
    city = city_match.group(1).strip() if city_match else None

    # Extract BHK (like 2BHK, 3 BHK)
    bhk_match = re.search(r'(\d+)\s*bhk', query, re.IGNORECASE)
    bhk = int(bhk_match.group(1)) if bhk_match else None

    # Extract budget (like under 1.2 Cr or below 80 Lakh)
    budget_match = re.search(r'under\s*₹?([\d\.]+)\s*(cr|crore|l|lakh)?', query, re.IGNORECASE)
    budget = None
    if budget_match:
        value = float(budget_match.group(1))
        unit = budget_match.group(2)
        if unit:
            if unit.lower().startswith('cr'):
                budget = value * 10000000  # Convert Cr to rupees
            elif unit.lower().startswith('l'):
                budget = value * 100000     # Convert Lakh to rupees
        else:
            budget = value

    return {
        'city': city,
        'bhk': bhk,
        'budget': budget or 0  # Default 0 if not found
    }
```

```python
import re
import pandas as pd

# 1️⃣ Extract filters from user query
def extract_filters(query):
    query = query.lower()
    filters = {'bhk': None, 'city': None, 'budget': 0}

    # Extract BHK
    bhk_match = re.search(r'(\d+)\s*bhk', query)
    if bhk_match:
        filters['bhk'] = int(bhk_match.group(1))

    # Extract budget (in lakhs or crores)
    budget_match = re.search(r'(\d+(\.\d+)?)\s*(cr|crore|l|lakh)', query)
    if budget_match:
        value = float(budget_match.group(1))
        if budget_match.group(3) in ['cr', 'crore']:
            filters['budget'] = value * 10000000
        else:
            filters['budget'] = value * 100000

    # Extract city (optional — example cities)
    city_match = re.search(r'in\s+([a-zA-Z\s]+)', query)
    if city_match:
        filters['city'] = city_match.group(1).strip().title()

    return filters


# 2️⃣ Search dataset for matching projects
def search_properties(filters, df):
    results = df.copy()

    # Filter by city (search in fullAddress)
    if filters['city']:
        results = results[results['fullAddress'].str.contains(filters['city'], case=False, na=False)]

    # Filter by BHK (check in projectCategory or summary)
    if filters['bhk']:
        results = results[results['projectCategory'].str.contains(str(filters['bhk']), case=False, na=False)]

    # Filter by budget (only if price exists)
    if 'price' in results.columns and filters['budget'] > 0:
        results = results[results['price'] <= filters['budget']]

    return results


# 3️⃣ Summarize results for chatbot reply
def summarize_results(filters, results):
    city = filters.get('city', 'your city')
    bhk = f"{filters.get('bhk')}BHK " if filters.get('bhk') else ''
    budget = filters.get('budget', 0)

    # Format readable budget
    budget_text = ''
    if budget > 0:
        if budget >= 10000000:
            budget_text = f"under ₹{budget/10000000:.1f} Cr"
        elif budget >= 100000:
            budget_text = f"under ₹{budget/100000:.1f} Lakh"

    # No results found
    if results.empty:
        if budget_text:
            return f"Sorry, no {bhk}properties found in {city} {budget_text}."
        else:
            return f"Sorry, no {bhk}properties found in {city}."

    # Success summary
    total = len(results)
    summary = f"Found {total} {bhk}projects in {city}"
    if budget_text:
        summary += f" {budget_text}"
    summary += "."

    summary += "\nHere are a few options:\n"
    if 'projectName' in results.columns:
        top_projects = results['projectName'].head(3).tolist()
        summary += " • " + "\n • ".join(top_projects)
    else:
        summary += "Project details unavailable."

    return summary


# 4️⃣ Final chatbot function
def property_bot(user_query, df):
    filters = extract_filters(user_query)
    results = search_properties(filters, df)
    response = summarize_results(filters, results)
    return response


# Example Usage
query = "3BHK flat in Pune under 1.2 Cr"
print(property_bot(query, project))
```
